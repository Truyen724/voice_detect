```python
import cv2
import librosa
import librosa.display
import numpy as np

# Đọc file âm thanh và tính toán Mel Spectrogram
y, sr = librosa.load("Data/Data/Test/CẨM LY/Ang-May-Buon-Cam-Ly/vocals.wav")
mel_spec = librosa.feature.melspectrogram(y=y, sr=sr)

# Chuyển đổi độ đo từ power sang decibel
mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)
mel_spec_db = np.uint8(255 * mel_spec_db / np.max(mel_spec_db))
cv2_imshow(mel_spec_db)
```

```python
import librosa
import os
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
src = "Data/Data/Test/"
for i in os.listdir(src):
    for j in os.listdir(src+i):
        s = src+i+"/"+ j
        if(s.endswith(".wav")==False):
          file_vocals  = s+"/" + "vocals.wav"
          y, sr = librosa.load(file_vocals)
          D = np.abs(librosa.stft(y))**2
          S = librosa.feature.melspectrogram(S=D)
          S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=512,
                                      fmax=8000)
          
          plt.figure(figsize=(10, 4))
          librosa.display.specshow(librosa.power_to_db(S,
                                                      ref=np.max),
                                  y_axis=None, fmax=8000,
                                x_axis=None)
          print(file_vocals)
          plt.axis = None
          plt.savefig(s+"/"+"melspectrogram.png",bbox_inches ="tight",dpi = 300 ,pad_inches = 0 )
          plt.tight_layout()
```

```python
import librosa
import os
import librosa.display
import numpy as np
import gc
import matplotlib.pyplot as plt
def generate_image(scr_wav, dest_wav):
      file_vocals  = scr_wav
      y, sr = librosa.load(file_vocals)
      D = np.abs(librosa.stft(y))**2
      S = librosa.feature.melspectrogram(S=D)
      S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=512,
                                  fmax=8000)
      plt.figure(figsize=(10, 4))
      librosa.display.specshow(librosa.power_to_db(S,
                                                  ref=np.max),
                              y_axis=None, fmax=8000,
                            x_axis=None)

      print(file_vocals)
      plt.axis = None
      plt.savefig(dest_wav,bbox_inches ="tight",dpi = 100 ,pad_inches = 0 )
      # Giải phóng bộ nhớ của biến y và sr
      del y, sr
      # Thu gom bộ nhớ
      gc.collect()

      del S
      # Thu gom bộ nhớ
      gc.collect()

      plt.figure().clear()
      plt.close()
      plt.cla()
      plt.clf()
      
  
```
